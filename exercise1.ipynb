# Q1: Reverse string + every 2nd character
def string_ops(s):
    reversed_str = s[::-1]
    second_chars = s[::2]
    return reversed_str, second_chars


# Q2: Squares of even numbers
def even_squares(lst):
    return [x*x for x in lst if x % 2 == 0]


# Q3: Character frequency
def char_frequency(s):
    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1
    return freq


# Q4: Indices of odd numbers
def odd_indices(lst):
    return [i for i, x in enumerate(lst) if x % 2 != 0]


# Q5: Rotate list right by k
def rotate_list(lst, k):
    k = k % len(lst)
    return lst[-k:] + lst[:-k]


# Q6: Two lists → dictionary
def lists_to_dict(keys, values):
    return dict(zip(keys, values))


# Q7: iter() and next() with exception handling
def manual_iterate(lst):
    it = iter(lst)
    while True:
        try:
            print(next(it))
        except StopIteration:
            break


# Q8: Flatten nested list
def flatten_list(nested):
    return [item for sublist in nested for item in sublist]


# Q9: Palindrome check
def is_palindrome(s):
    return s == s[::-1]


# Q10: Second largest (without sorting)
def second_largest(lst):
    first = second = float('-inf')
    for num in lst:
        if num > first:
            second = first
            first = num
        elif num > second and num != first:
            second = num
    return second
